```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 2 time series
t = np.linspace(1, 9, 9)
u1 = np.array([1.,1.,1.,2.,2.,2.,3.,3.,3.])
print "u1", u1
print "u1, mean, std", u1.mean(), u1.std()
u2 = np.array([1.,1.,1.,2.,2.,2.,4.5,2.25,2.25])
print "u2 mean and std", u2.mean(), u2.std()
plt.plot(t, u1)
plt.plot(t, u2)
plt.title('u vs t');
```
